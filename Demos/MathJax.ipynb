```python
import matplotlib.pyplot as plt

# Define the size of the canvas (in inches)
canvas_width = 12
canvas_height = 8

# Define the size of the smaller figure (in inches)
fig_width = 6
fig_height = 4

# Create a larger canvas
fig_canvas = plt.figure(figsize=(canvas_width, canvas_height))

# Add a smaller figure (axes) centered within the canvas
left = (canvas_width - fig_width) / (2 * canvas_width)
bottom = (canvas_height - fig_height) / (2 * canvas_height)
width = fig_width / canvas_width
height = fig_height / canvas_height

# Add the smaller axes
ax = fig_canvas.add_axes([left, bottom, width, height])

# Plot some data on the smaller axes
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
ax.plot(x, y)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Smaller Figure Centered on Larger Canvas')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Ensure matplotlib inline for Jupyter Notebook
%matplotlib inline

# Define the size of the figure (in inches)
fig_width = 12
fig_height = 8

# Define the relative size of the axes within the figure
axes_width_rel = 0.6
axes_height_rel = 0.4

# Define the position of the axes (relative to figure size)
left_rel = 0.2
bottom_rel = 0.3

# Create the figure
fig = plt.figure(figsize=(fig_width, fig_height))

# Add axes with specified size and position relative to figure dimensions
ax = fig.add_axes([left_rel, bottom_rel, axes_width_rel, axes_height_rel])

# Example data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot some data
ax.plot(x, y)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Axes Size Relative to Figure Dimensions')

# Adjust layout to prevent autoscaling issues
fig.subplots_adjust(left=0, right=1, bottom=0, top=1)  # Adjust this as needed

# Optionally, set aspect ratio if necessary
ax.set_aspect('equal', adjustable='box')

# After plotting, adjust the margins
plot_margin = 0.25

# Get current axis limits
x0, x1, y0, y1 = plt.axis()

# Adjust axis limits to add padding
plt.axis((x0 - plot_margin,
          x1 + plot_margin,
          y0 - plot_margin,
          y1 + plot_margin))

# Re-show the plot with adjusted margins
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Ensure matplotlib inline for Jupyter Notebook
%matplotlib inline

# Define the size of the figure (in inches)
fig_width = 12
fig_height = 8

# Create the figure
fig = plt.figure(figsize=(fig_width, fig_height), linewidth=1, edgecolor='black')

# Add axes to the figure
ax = fig.add_axes([0.2, 0.3, 0.6, 0.4])  # Position and size of axes

# Example data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot some data
ax.plot(x, y)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Using plt.margins() to Adjust Margins')
ax.margins(x=0.3, y=0.3)

# Set margins as a fraction of the data range
plt.margins(x=0, y=0)  # Adds 10% margin on both sides of x and y axes

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

# Create the figure
fig = plt.figure(figsize=(10, 10))

# Create an 8x8 grid of subplots
gs = gridspec.GridSpec(8, 8)

# Define the grid positions for the middle 4 grid boxes
# This spans rows 2-5 and columns 2-5
ax = fig.add_subplot(gs[2:6, 2:6])

# Example data for the centered plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot some data
ax.plot(x, y)
ax.set_title('Plot Spanning Middle 4 Grid Boxes')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Show the plot
plt.show()
```
